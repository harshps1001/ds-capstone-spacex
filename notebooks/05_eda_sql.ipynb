```python
import pandas as pd
import sqlite3

df = pd.read_csv('../data/spacex_launch_dash.csv')
conn = sqlite3.connect('spacex.db')
df.to_sql('SPACEXTBL', conn, if_exists='replace', index=False)
print('Database created: spacex.db')
```

## Q1 — Unique Launch Sites

```python
q1 = pd.read_sql('SELECT DISTINCT "Launch_Site" FROM SPACEXTBL', conn)
print('Unique launch sites:')
print(q1)
```

## Q2 — Launches from CCAFS

```python
q2 = pd.read_sql("SELECT * FROM SPACEXTBL WHERE Launch_Site LIKE 'CCA%' LIMIT 5", conn)
print(q2[['FlightNumber','Date','BoosterVersion','PayloadMass','Orbit','Class']])
```

## Q3 — Total Payload (NASA CRS)

```python
q3 = pd.read_sql("SELECT SUM(PayloadMass) AS total FROM SPACEXTBL WHERE Customer='NASA (CRS)'", conn)
print(f'Total NASA CRS payload: {q3["total"][0]} kg')
```

## Q9 — Drone Ship Boosters (5000–15000 kg payload)

```python
q9 = pd.read_sql("""
    SELECT BoosterVersion, PayloadMass, LaunchSite, Outcome
    FROM SPACEXTBL
    WHERE LandingPad IS NOT NULL
    AND PayloadMass BETWEEN 5000 AND 15000
""", conn)
print(q9)
```

## Q10 — Landing Outcome Counts

```python
q10 = pd.read_sql("""
    SELECT Outcome, COUNT(*) AS Count
    FROM SPACEXTBL
    GROUP BY Outcome
    ORDER BY Count DESC
""", conn)
print(q10)
conn.close()
```
