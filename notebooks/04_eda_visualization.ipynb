```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df = pd.read_csv('../data/spacex_launch_dash.csv')
df.head()
```

## Scatter: Flight Number vs Launch Site

```python
fig, ax = plt.subplots(figsize=(12,5))
sns.scatterplot(x='FlightNumber', y='LaunchSite', hue='Class',
                data=df, palette={0:'red',1:'green'}, ax=ax)
ax.set_title('Flight Number vs Launch Site (Green=Success, Red=Failure)')
ax.set_xlabel('Flight Number')
ax.set_ylabel('Launch Site')
plt.tight_layout()
plt.savefig('eda_flight_vs_site.png', dpi=150)
plt.show()
```

## Scatter: Payload Mass vs Launch Site

```python
fig, ax = plt.subplots(figsize=(12,5))
sns.scatterplot(x='PayloadMass', y='LaunchSite', hue='Class',
                data=df, palette={0:'red',1:'green'}, ax=ax)
ax.set_title('Payload Mass vs Launch Site')
ax.set_xlabel('Payload Mass (kg)')
plt.tight_layout()
plt.savefig('eda_payload_vs_site.png', dpi=150)
plt.show()
```

## Bar Chart: Success Rate by Orbit Type

```python
orbit_success = df.groupby('Orbit')['Class'].mean().sort_values()
fig, ax = plt.subplots(figsize=(10,5))
orbit_success.plot(kind='barh', color='steelblue', ax=ax)
ax.set_title('Landing Success Rate by Orbit Type')
ax.set_xlabel('Success Rate')
ax.set_xlim(0, 1.1)
for i, v in enumerate(orbit_success):
    ax.text(v+0.01, i, f'{v:.0%}', va='center')
plt.tight_layout()
plt.savefig('eda_orbit_success.png', dpi=150)
plt.show()
```

## Line Chart: Annual Success Rate Trend

```python
df['Year'] = pd.to_datetime(df['Date']).dt.year
year_success = df.groupby('Year')['Class'].mean() * 100
fig, ax = plt.subplots(figsize=(10,4))
year_success.plot(kind='line', marker='o', color='dodgerblue', linewidth=2.5, ax=ax)
ax.set_title('Annual Landing Success Rate (%) — 2013 to 2020')
ax.set_ylabel('Success Rate (%)')
ax.set_ylim(0, 110)
for yr, val in year_success.items():
    ax.annotate(f'{val:.0f}%', (yr, val), textcoords='offset points', xytext=(0,8))
plt.tight_layout()
plt.savefig('eda_yearly_trend.png', dpi=150)
plt.show()
```
